```python
#checking stationarity, using the rolling statistics plots along with Dickey-Fuller test

from statsmodels.tsa.stattools import adfuller
def test_stationarity(fdf):
    
#Determing rolling statistics
    rolmean = fdf.rolling(window=12).mean()
    rolstd = fdf.rolling(window=12).std()
    print(rolmean,rolstd)
#Plot rolling statistics:
    orig = plt.plot(fdf, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
    
```

```python
#Perform Dickey-Fuller test:
print('Results of Dickey-Fuller Test:')
dftest = adfuller(fdf['India_test_case'], autolag='AIC')
dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
print(dfoutput)
```

```python
#ACF and PACF plots:
from statsmodels.tsa.stattools import acf, pacf
fdf_log_diff = fdf_log_diff.replace(np.inf, np.nan).replace(-np.inf, np.nan).dropna()
```

```python
lag_acf = acf(fdf_log_diff, nlags=20)
lag_pacf = pacf(fdf_log_diff, nlags=20, method='ols')
```

```python
#Plot ACF: 
plt.subplot(121) 
plt.plot(lag_acf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(fdf_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(fdf_log_diff)),linestyle='--',color='gray')
plt.title('Autocorrelation Function')
```

```python
#Plot PACF:
plt.subplot(122)
plt.plot(lag_pacf)
plt.axhline(y=0,linestyle='--',color='gray')
plt.axhline(y=-1.96/np.sqrt(len(fdf_log_diff)),linestyle='--',color='gray')
plt.axhline(y=1.96/np.sqrt(len(fdf_log_diff)),linestyle='--',color='gray')
plt.title('Partial Autocorrelation Function')
plt.tight_layout()
```
